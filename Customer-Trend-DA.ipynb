```python
#Loading the dataset using pandas

import pandas as pd

df = pd.read_csv('customer_shopping_behavior.csv')
```

```python
df.head()
```

```python
# All info about data set
df.info()
```

```python
# Summary statistics using .describe() , include all -> includes non numerical value as well
df.describe(include='all')
```

```python
# check missing values
df.isna().sum()
```

```python
# replace missing values by mean or median , we chose median bcqz mean is affected by outliers
# rating vary with category 
# not to replace all missing value with overall median
# we will impute median within each category 

df['Review Rating'] = df.groupby('Category')['Review Rating'].transform(lambda x: x.fillna(x.median()))
```

```python
df.isna().sum()
```

```python
df.columns = df.columns.str.lower()
df.columns = df.columns.str.replace(' ','_')
df = df.rename(columns={'purchase_amount_(usd)':'purchase_amount'})
```

```python
df.columns
```

```python
# Create a column age_group
labels = ['Young Adult','Adult','Middle-aged','Senior']
df['age_group'] = pd.qcut(df['age'],q=4, labels=labels)
```

```python
df[['age','age_group']].head(10)  
```

```python
# create purchase_frequncy_days
frequency_mapping = {
    'Fortnightly': 14,
    'Weekly': 7,
    'Monthly': 30,
    'Quarterly': 90,
    'Bi-Weekly': 14,
    'Annually': 365,
    'Every 3 Months': 90
}
df['purchase_frequency_days'] = df['frequency_of_purchases'].map(frequency_mapping)
```
